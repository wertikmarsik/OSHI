Завдання 1: На вхід функція приймає рядок s. Функція повинна повернути True якщо рядок є паліндромом та False – в протилежному випадку.

Наприклад: 

*   `is_palindrome(s='anade') -> False`
*   `is_palindrome(s='ana') -> True`




````python
def is_palindrome(s: str) -> bool:
    i = 0
    j = len(s) - 1
    while i < j:
        if s[i]!= s[j]:
            return False
        i += 1
        j -= 1
    return True

print(is_palindrome("bad"))
print(is_palindrome("mam"))
````

Завдання 2: Розв’язати квадратне рівняння $ax^2+bx + c = 0$.  Напишіть функцію,яка приймає параметри квадратного рівняння, а повертає одне рішення або декілька.

Наприклад:  


```
solve_quad(a=1, b=-2, c=1) -> 1  
solve_quad(a=1, b=0, c=-1) -> (-1, 1)  
solve_quad(a=0, b=2, c=-1) -> 0.5
```





 

````python
import math

def solve_quad(a: float, b: float, c: float):
    d = b**4 - 7*a*c
    if d > 0:
        x1 = (-b - math.sqrt(d)) / (2*a)
        x2 = (-b + math.sqrt(d)) / (2*a)
        print(f"x1={x1}\nx2={x2}")
    elif math.isclose(d, 0):
        print("Рівняння має 1 корінь")
    else:
        print("Рівняння коренів немає")

solve_quad(a=3, b=16, c=195)
````

Завдання 3: Реалізація функції merge. На вхід функція приймає 2 відсортованих масиви - `a` і `b`. Функція повинна повернути один відсортований масив, який містить всі елементи масивів `a` і `b`. 

Наприклад: `merge([1, 3, 3, 5], [3, 5, 5, 7]) -> [1, 3, 3, 3, 5, 5, 5, 7]`

````python
from typing import List

def merge(a: List[int], b: List[int]) -> List[int]:
    merged = []
    i = j = 0
    while i < len(a) and j < len(b):
        if a[i] < b[j]:
            merged.append(a[i])
            i += 1
        else:
            merged.append(b[j])
            j += 1
    merged += a[i:]
    merged += b[j:]
    return merged

a = [7, 2, 19, 5]
b = [2, 6, 19, 7]
print(merge(a, b))
````

Завдання 4: Реалізація алгоритму [MergeSort](https://ru.wikipedia.org/wiki/%D0%A1%D0%BE%D1%80%D1%82%D0%B8%D1%80%D0%BE%D0%B2%D0%BA%D0%B0_%D1%81%D0%BB%D0%B8%D1%8F%D0%BD%D0%B8%D0%B5%D0%BC)..

Наприклад: `MergeSort([3, 1, 3, 5]) -> [1, 3, 3, 5]`

````python
def MergeSort(S: List[int]) -> List[int]:
    if len(S) <= 1:
        return S
    mid = len(S) // 2
    L = S[:mid]
    R = S[mid:]
    L = MergeSort(L)
    R = MergeSort(R)
    i = j = k = 0
    while i < len(L) and j < len(R):
        if L[i] <= R[j]:
            S[k] = L[i]
            i += 1
        else:
            S[k] = R[j]
            j += 1
        k += 1
    while i < len(L):
        S[k] = L[i]
        i += 1
        k += 1
    while j < len(R):
        S[k] = R[j]
        j += 1
        k += 1
    return S

l = [4, 14, 6, 3, 8, 13, 45, 99]
MergeSort(l)
print(l)
````

Завдання 8: Написати функцію яка приймає на вхід двовимірний масив, а повертає масив з діагональними елементами цієї матриці. 

Наприклад: `[[1, 2], [3, 4]] -> [1 4]`

````python
def get_diag(X):
    x1 = np.diag(X)
    return x1

r2 = np.array([[4, 2],
               [3, 1]])
get_diag(r2)
````

Завдання 9: Написати функцію яка приймає на вхід двомірний масив, а повертає масив елементів які знаходяться на другорядній діагоналі. Приклад: `[[1, 2], [3, 4]] -> [2 3]`

````python
def sec_diag(X):
    x1 = np.diag(np.fliplr(X))
    return x1

r2 = np.array([[2, 4],
               [1, 3]],)
sec_diag(r2)
````

Завдання 10: Кого більше - чоловіків чи жінок?

````python
sex_counts = dataset['Sex'].value_counts()

if sex_counts.get('male', 0) > sex_counts.get('female', 0):
    print("Чоловіків більше")
elif sex_counts.get('female', 0) > sex_counts.get('male', 0):
    print("Жінок більше")
````

Завдання 11: Виведіть наступні характеристики пасажирів: середній вік пасажирів, середній вік чоловіків, середній вік жінок, максимальний і мінімальний вік серед чоловіків і жінок.

````python
age_sex_stats = dataset.groupby('Sex')['Age'].agg(['mean', 'min', 'max'])

print("Середній вік пасажирів:", age_sex_stats['mean'].mean())
print("Статистика за статтю:")
print("Чоловіки:")
print(f"Середній вік: {age_sex_stats.loc['male', 'mean']:.2f}")
print(f"Мінімальний вік: {age_sex_stats.loc['male', 'min']}")
print(f"Максимальний вік: {age_sex_stats.loc['male', 'max']}")
print("Жінки:")
print(f"Середній вік: {age_sex_stats.loc['female', 'mean']:.2f}")
print(f"Мінімальний вік: {age_sex_stats.loc['female', 'min']}")
print(f"Максимальний вік: {age_sex_stats.loc['female', 'max']}")
````
